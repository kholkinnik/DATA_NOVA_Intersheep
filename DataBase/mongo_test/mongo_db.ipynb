```python
# –í—ã—á–∏—Å–ª—è—é —Ä–∞–∑–Ω–∏—Ü—É —Ç–µ–∫—É—â–µ–π –¥–∞—Ç—ã –∏ 30 –¥–Ω–µ–π, –¥–ª—è —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏–∏ –ø–æ –¥–∞—Ç–µ —Ä–µ–≥–∏—Å—Ç—Ä–∞—Ü–∏–∏
dif_date_reg = datetime.now() - timedelta(days = 30)

# –í—ã—á–∏—Å–ª—è—é —Ä–∞–∑–Ω–∏—Ü—É —Ç–µ–∫—É—â–µ–π –¥–∞—Ç—ã –∏ –∏ 14 –¥–Ω–µ–π, –¥–ª—è —Ñ–∏–ª—å—Ç—Ä–∞—Ü–∏–∏ –±–µ–∑–¥–µ–π—Å—Ç–≤–∏—è –ø–æ–ª—å–∑–æ–≤—Ç–∞–µ–ª—è –≤ —Ç–µ—á–µ–Ω–∏–∏ 14 –¥–Ω–µ–π
dif_date_event = datetime.now() - timedelta(days = 14) 
```

```python
# —Ñ–æ—Ä–º–∏—Ä—É—é –∑–∞–ø—Ä–æ—Å –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –∑–∞–¥–∞–Ω–∏—è
query = {
    "$and": [
        {'user_info.registration_date': {'$lt': dif_date_reg}},
        {'event_time': {'$lt': dif_date_event}}
    ]   
}
```

```python
#–≤—ã–ø–æ–ª–Ω—è—é –∑–∞–ø—Ä–æ—Å –Ω–∞ –≤ mongo
archived_users = collection.find(query, {"user_id": 1, "_id": 0})

# –°–æ–∑–¥–∞—ë–º —Å–ø–∏—Å–æ–∫ user_id –¥–ª—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏ –≤ –æ—Ç—á–µ—Ç–µ
archived_user_ids = [doc['user_id'] for doc in archived_users]
```

```python
#–§–æ—Ä–º–∞—Ç –æ—Ç—á–µ—Ç–∞ 
report = {
    "date" : datetime.now().date().strftime("%Y-%m-%d"),
    "archived_users_count": len(archived_user_ids),
    "archived_user_ids" : archived_user_ids
}
```

```python
# –ü—É—Ç—å –¥–ª—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω–∏—è –æ—Ç—á—ë—Ç–∞
output_dir = os.getcwd()
file_path = os.path.join(output_dir, f'''report_from_{datetime.now().date().strftime("%Y-%m-%d")}.json''')
```

```python
# –ó–∞–ø–∏—Å—å –æ—Ç—á—ë—Ç–∞ –≤ —Ñ–∞–π–ª .json
with open(file_path, 'w', encoding='utf-8') as f:
    json.dump(report, f, ensure_ascii=False, indent=2)

print(f"–û—Ç—á—ë—Ç —Å–æ—Ö—Ä–∞–Ω—ë–Ω –≤ {file_path}")
```
